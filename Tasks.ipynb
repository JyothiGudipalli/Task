```python
import numpy as np
```

```python
a=np.random.randint(5,size=(5,3))
print(a)
```

```python
import matplotlib.pyplot as plt

plt.hist(a)
plt.title("Histogram for 2D array")
plt.show()
```

```python
b=np.random.randint(5,size=(4,4))
x=b[1]
y=b[2]
```

```python
plt.hist2d(x,y)
plt.title("histogram for 2d")
plt.show()
```
